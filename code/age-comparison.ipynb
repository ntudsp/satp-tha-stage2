```python
from scipy.stats import kstest
import pandas as pd
import os
```

```python
df = pd.read_excel(os.path.join('..','data','SATP Dataset v1.4.xlsx'))
```

```python
df
```

```python
df = df[~df['Language'].isin(['nld','vie'])]
df['Age'] = df['Age'].astype(float)
df
```

```python
langs = []
pids = []
ages = []
for (lang, pid), df_sub in df.groupby(by=['Language', 'Participant']):
    try:
        langs.append(lang)
        pids.append(pid)
        ages.append(df_sub['Age'].mean())
    except:
        print('one error')
df1 = pd.DataFrame({'lang':langs, 'pid':pids, 'age':ages})
df1
```

```python
df1.groupby('lang').describe()
```

```python
df2 = pd.read_excel('..\\data\\[SATP Thai Stage 2] Demographic information questionnaire (Responses).xlsx')
df2 = df2[~df2['Participant ID'].isin([7,27])]
df2
```

```python
for lang, df_sub in df1.groupby('lang'):
    print(lang, df_sub.shape)
    if df_sub['age'].dropna().empty:
        print('\tAll NA')
    else:
        K = kstest(df_sub['age'].dropna(),df2['What is your age?'])
        print(f'\t{K}')
```
